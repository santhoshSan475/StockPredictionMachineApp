```python
##SUBPLOT & LEGEND
```

```python
days = [1,2,3,4,5]
original_price = [15,21,28,35,45]
predicted_price = [18,23,28,37,43]

plt.plot(days, original_price, 'b-o',label='original');
plt.plot(days,  predicted_price, 'r-o', label='predicted');
plt.title('prediction price');
plt.xlabel('days');
plt.ylabel('price');
plt.legend();
```

```python
plt.subplot(3,1,1);
plt.plot(days, original_price, 'b-o',label='original');
plt.title('prediction price');
plt.xlabel('days');
plt.ylabel('price');
plt.legend();



plt.subplot(3,1,2)
plt.plot(days, predicted_price, 'r-o',label='predicted');
plt.title('prediction price');
plt.xlabel('days');
plt.ylabel('price');
plt.legend();

plt.subplot(3,1,3)

difference = [ o-p for o,p in zip(original_price,predicted_price) ]
plt.bar(days, difference, color = 'purple',label='difference');
plt.title('prediction price');
plt.xlabel('days');
plt.ylabel('price');
plt.legend();
```
